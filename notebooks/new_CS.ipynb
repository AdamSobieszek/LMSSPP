## 1. Core backend smoke tests

```python
%cd /Users/adamsobieszek/pitch/pitch-website/public/notebooks/kuramoto/LMSSPP/src
from lmsspp.core.cucker_smale import (
    simulate_cs_widget_trajectory,
    compare_reduced_vs_second_order_cs,
)

N, d = 64, 3
weights = torch.ones(N, dtype=torch.float64) / N
x0 = torch.randn(N, d, dtype=torch.float64) * 0.2
omega = torch.randn(N, d, dtype=torch.float64) * 0.05

traj = simulate_cs_widget_trajectory(
    x0=x0,
    omega=omega,
    weights=weights,
    alpha=0.35,
    eps=0.02,
    coupling=1.0,
    dt=1e-2,
    steps=120,
)

traj.w.shape, traj.x_lab.shape if traj.x_lab is not None else None
```

```python
v0 = torch.randn(N, d, dtype=torch.float64) * 0.05
diag = compare_reduced_vs_second_order_cs(
    x0=x0,
    v0=v0,
    weights=weights,
    alpha=0.35,
    eps=0.02,
    coupling=1.0,
    dt=1e-3,
    steps=20,
)

{
    'x_rms_final': float(diag['x_rms'][-1]),
    'v_rms_final': float(diag['v_rms'][-1]),
    'omega_inv_max': float(diag['omega_inv_max'].max()),
}
```
